```python
import subprocess
import sys

libraries = [
    'transformers>=4.30.0',
    'torch>=2.0.0',
    'pillow>=10.0.0',
    'opencv-python>=4.7.0',
    'matplotlib>=3.7.0',
    'requests>=2.31.0',
    'numpy>=1.24.0'
]

print('📦 Libraries install ho rahi hain...\n')
for lib in libraries:
    result = subprocess.run(
        [sys.executable, '-m', 'pip', 'install', lib, '-q'],
        capture_output=True, text=True
    )
    status = '✅' if result.returncode == 0 else '❌'
    print(f'{status} {lib}')

print('\n🎉 Installation complete!')
```

```python
import torch
import transformers
import PIL
import cv2
import matplotlib
import numpy as np
import requests

print('=' * 55)
print('   🔍 ENVIRONMENT VERIFICATION REPORT')
print('=' * 55)

versions = {
    'PyTorch':          torch.__version__,
    'Transformers':     transformers.__version__,
    'Pillow (PIL)':     PIL.__version__,
    'OpenCV':           cv2.__version__,
    'Matplotlib':       matplotlib.__version__,
    'NumPy':            np.__version__,
    'Requests':         requests.__version__,
    'Python':           __import__('sys').version.split()[0]
}

for lib, ver in versions.items():
    print(f'  ✅ {lib:<20} → v{ver}')

print('=' * 55)
```

```python
import torch

print('=' * 55)
print('   ⚡ HARDWARE CONFIGURATION REPORT')
print('=' * 55)

# CUDA (NVIDIA GPU) check
cuda_available = torch.cuda.is_available()
device = 'cuda' if cuda_available else 'cpu'

if cuda_available:
    print(f'  🚀 GPU DETECTED — CUDA Enabled!')
    print(f'  📌 GPU Name    : {torch.cuda.get_device_name(0)}')
    print(f'  📌 GPU Memory  : {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB')
    print(f'  📌 CUDA Version: {torch.version.cuda}')
else:
    print('  🖥️  CPU MODE — No GPU detected')
    print('  ℹ️  Model CPU par bhi chalega, thoda slow hoga')

print(f'\n  🎯 Active Device: {device.upper()}')
print('=' * 55)
print('\n✅ Environment ready hai! Notebook 02 kholo.')
```
